```python
import pandas as pd
```

```python
df = pd.read_csv("./rankings.csv")

for index, row in df.iterrows():
  df.loc[index, 'Member Name'] = row['Member Name'].lower().strip()
  df.loc[index, 'Group Name'] = row['Group Name'].lower().strip()
```

```python
rankings = {1: 90, 2: 83, 3: 75, 4: 63, 5: 54, 6: 49, 7: 42, 8: 35, 9: 32, 10: 26, 11: 24, 12: 20, 13: 16, 14: 12, 15: 10, 16: 8, 17: 6, 18: 4, 19: 2, 20: 1}
```

```python
def get_rankings(items):
    points = rankings[int(items["Rank"])]
    name = f'{items["Member Name"]}, {items["Group Name"]}'

    if name in calculated:
        calculated[name]["points"] += points
    else:
        calculated[name] = {"points": points}
```

```python
calculated = {}
for index, row in df.iterrows():
    get_rankings(row)
```

```python
final = {}
for k, v in calculated.items():
    final[k] = v["points"]
  
final_list = sorted(final.items(), key=lambda x: x[1], reverse=True)
top_20 = final_list[:20]
```
